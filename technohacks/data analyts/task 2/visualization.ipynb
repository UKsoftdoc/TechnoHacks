```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<font size="4"> **Loading Dataset**</font>


```python
Iris_data=pd.read_csv("Iris.csv")
```

```python
Iris_data.head()
```

<font size="4"> **Removing Unwanted columns**</font>

```python
Iris_data.duplicated().sum()
```

```python
#removing id attribute because its not important for classification
Iris_data.drop("Id", axis=1,inplace=True)
```

```python
Iris_data.head()
```

<font size="4"> **Visualization**</font>

<font size="4"> Visualize SepalLength column</font>

```python
plt.hist(Iris_data["SepalLengthCm"],color="yellow")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Iris data Sepal Length")
plt.show()
```

<font size="3"> **Visualize SepalWidth column**</font>

```python
plt.hist(Iris_data["SepalWidthCm"],color="navy",alpha=0.7)
plt.xlabel("Sepal Width")
plt.ylabel("Frequency")
plt.title("Histogram of Iris data Sepal Width")
plt.show()
```

<font size="3"> **Visualize PetalLength column**</font>

```python
plt.hist(Iris_data["PetalLengthCm"],color="orange")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title("Histogram of Iris data Petal Length")
plt.show()
```

<font size="3"> **Visualize PetalWidth column**</font>

```python
plt.hist(Iris_data["PetalWidthCm"],color="orange")
plt.xlabel("Petal Width")
plt.ylabel("Frequency")
plt.title("Histogram of Iris data Petal Width")
plt.show()
```

<font size="3"> **Visualize Species column**</font>

```python
plt.hist(Iris_data["Species"],color="black")
plt.xlabel("Species")
plt.ylabel("Frequency")
plt.title("Histogram of Iris data Species")
plt.show()
```
